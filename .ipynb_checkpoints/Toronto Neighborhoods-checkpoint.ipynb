#### First Exercise


Importing Necesary Libraries for the Project

```python
import pandas as pd
import numpy as np
```

Scrapping data for the Wikipedia url given and using only the first table received:

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url)[0]
df
```

- Getting rid of all rows containing 'Not assigned'
- Fixing the indices
- Printing the shape of the resulting dataframe

```python
df = df[df['Borough']!="Not assigned"]
df = df.reset_index(drop=True)
df
np.shape(df)
```

#### Second Exercise

Loading the data from the csv file:

```python
csv = 'http://cocl.us/Geospatial_data'
geo = pd.read_csv(csv)
geo
```

Merging the two dataframes:

```python
toronto_df = pd.merge(df,geo).sort_values(by=["Postal Code"])
toronto_df
```
